<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df.dropna(subset=['AVG_READING_4_SCORE'], inplace=True)
df['PORTION_OF_TOTAL_SPENDING'] = df['INSTRUCTION_EXPENDITURE'] / df['TOTAL_EXPENDITURE']
df['INSTRUCTION_EXPENDITURE_PER_STUDENT'] = df['INSTRUCTION_EXPENDITURE'] / df['GRADES_ALL_G']
```

Feature engineering justification: First, I decided to clean up the dataset by removing any missing variables because they would not be able to contribute to the data and models. Second, I created a column to see how much of the total expenditure was being spent on the children and instruction. Third, I created an expenditure per student column to see how equitable the spending was in the state.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['PORTION_OF_TOTAL_SPENDING','INSTRUCTION_EXPENDITURE_PER_STUDENT']].dropna()
y = df.loc[X.index]['AVG_READING_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
      X, y, test_size=.4, random_state=42)
```
